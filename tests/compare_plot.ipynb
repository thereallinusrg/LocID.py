```python
matches = df[cols]
matches
```

```python
def set_count(set_locid):
    set_n = len(set_locid[1])-1
    return set_n

set_numbers = pd.DataFrame()

for column in cols:
    set_name = column.replace("match", "SeT_Lvl")
    set_numbers[set_name] = matches[column].apply(set_count)

set_numbers
```

```python
res = pd.DataFrame()
cols = matches.columns
for n in range(0, len(matches.columns)):
    res[f"{cols[n]}"] = df.apply(lambda x: x[f"{cols[n]}"][0], axis=1)
```

Sort by lvl / tol 

```python
lvlID = []
tolID = []
for mID in res.columns:
    tolID.append(float(mID.split('_')[1][1:]))
    lvlID.append(mID.split('_')[2][1:])
```

```python
tuples = zip(tolID, lvlID)
index = pd.MultiIndex.from_tuples(tuples, names=["tol", "lvl"])
```

```python
res.columns = index

res
```

```python
#deltas over tol
delta_tol = res.groupby(level=1, axis=1).diff()
delta_tol
```

```python
#deltas over lvl
delta_lvl = res.groupby(level=0, axis=1).diff()
delta_lvl
```

# Plots

```python
def t_table(df):
    df = df.dropna(axis=1, how='any')
    dataBig = pd.DataFrame()
    for test in range(0,len(df)):
        data = df.filter(items=[test], axis=0)
        dataT = data.T
        dataT.reset_index(level=['tol', 'lvl'], inplace=True)
        dataT['test'] = test
        dataT.columns=['tol', 'lvl', 'val', 'test']
        dataBig = pd.concat([dataBig, dataT], axis=0)
        dataBig['val'] = dataBig['val'].abs()
    return dataBig
```

```python
dt_tol = t_table(delta_tol.reset_index())
dt_lvl = t_table(delta_lvl.reset_index())
```

```python
dt_lvl
```

```python
# LEVEL
dataBig = dt_lvl.drop(index=0)
sns.set_theme(style="whitegrid")

# Initialize a grid of plots with an Axes for each walk
grid = sns.FacetGrid(dataBig, col="tol", col_wrap=3, hue="test", 
                    palette="tab10", height=4, col_order=[0.001, 0.0001, 1e-05, 1e-06, 1e-07]
                    )
                    

# Draw a line plot to show the trajectory of each test
grid.map(plt.plot, "lvl", "val", marker="o", color=(0.1,0.1,0.5,0.1))     

# Adjust the tick positions and labels
#grid.set(ylim=(0, 1)) 

# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=3)

#grid.add_legend(legend_data=None, title=None, label_order=None, adjust_subtitles=False)

grid.fig.suptitle('[lvl] Deltas of match values, differences between levels', y=1.05)
```

```python
# TOLERANCE
dataBig = dt_tol.drop(index=0)
sns.set_theme(style="whitegrid")

# Initialize a grid of plots with an Axes for each walk
grid = sns.FacetGrid(dataBig, col="lvl", col_wrap=3, hue="test", 
                    palette="tab10", height=4
                    ) 

# Draw a line plot to show the trajectory of each test
grid.map(plt.plot, "tol", "val", marker="o", color=(0.1,0.1,0.5,0.1))

# Adjust the tick positions and labels
grid.set(xlim=(1e-04, 1e-07), ylim=(0, 1))

# Adjust the arrangement of the plots
grid.fig.tight_layout(w_pad=3)

# Set the x-Axis to log scale
grid.set(xscale="log")

#grid.add_legend(legend_data=None, title=None, label_order=None, adjust_subtitles=False)

grid.fig.suptitle('[tol] Deltas of match value, between tolerances', y=1.05)
```
